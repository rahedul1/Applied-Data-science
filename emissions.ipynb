```python
import pandas as pd 
import matplotlib.pyplot as plt 
def Emission(P):#emission function
      x=P['Country'] #on x-axis is the year
      y=P['Amount'] #on y-axis is the amount
      plt.figure(figsize=(22,5))#setting the size of the figure
      plt.bar(x,y,color="Blue") #figure is plotted with blue bars
      plt.title('CO2 Emissions for Different Countries in 2017') #title of our chart
      plt.xlabel('Countries') #x label
      plt.ylabel('Amount of CO2') #y label
      plt.show()
def Emission18(P):#emission function
      x=P['Country'] #on x-axis is the year
      y=P['Amount'] #on y-axis is the population
      plt.figure(figsize=(22,5))#setting the size of the figure
      plt.bar(x,y,color="Blue") #figure is plotted with blue bars
      plt.title('CO2 Emissions for Different Countries in 2018') #this is title of our chart
      plt.xlabel('Countries') #x label
      plt.ylabel('Amount of CO2') #y label in our figure
      plt.show()

def Temp17(M):
      x=M['AvgTemperature']#on x-axis is the year
      y=M['Country']#on y-axis is the value
      M.plot.box(grid='True')
      plt.title('Average Temperature in Year 2017')#title of figure
      plt.xlabel('Temperature')#x label
      plt.ylabel('Country')#y label
      plt.show()

def Temp18(B):
      x=B['AvgTemperature']#on x-axis is the year
      y=B['Country']#on y-axis is the value
      plt.scatter(x,y,color="Purple")
      plt.title('Average Temperature in Year 2018')#title of our chart
      plt.xlabel('Temperature')#x label
      plt.ylabel('Country')#y label
      plt.show()

A1=pd.read_excel('emission.xlsx','CO17')#reading emission  in year 2017
A12=pd.read_excel('emission.xlsx','CO18')#reading emission  in year 2018
T1=pd.read_excel('emission.xlsx','Temp17')#reading temp data of year 2017
T12=pd.read_excel('emission.xlsx','Temp18')#reading temp data of year 2018
Emission(A1)#function called to draw CO2 trend
Emission18(A12)#function called to draw CO2 trend
Temp17(T1)#function to draw temperature 2017
Temp18(T12)#function to draw temperature 2018
```
